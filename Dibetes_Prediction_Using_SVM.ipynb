```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import warnings
```

```python
diabetes_dataset = pd.read_csv("diabetes.csv")
```

```python
X = diabetes_dataset.drop(columns = "Outcome", axis = 1)
Y = diabetes_dataset["Outcome"]
```

```python
print(X)
```

```python
print(Y)
```

```python
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset["Outcome"]
```

```python
print(X)
print(Y)
```

```python
warnings.filterwarnings("ignore", message=".*X does not have valid feature names.*")
```

```python
[X_train , X_test ,Y_train , Y_test] = train_test_split(X,Y,train_size = 0.8, stratify = Y,random_state=1)
print(X.shape,X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel = "linear")
classifier.fit(X_train,Y_train)
```

```python
X_train_data_predict = classifier.predict(X_train)
X_train_data_accuracy = accuracy_score(X_train_data_predict,Y_train)
print("SVM Model accuracy on Training data is :", X_train_data_accuracy)
```

```python
X_test_data_prediction = classifier.predict(X_test)
X_test_data_accuracy = accuracy_score(X_test_data_prediction,Y_test)
print("SVM Model accuracy on Test data is :", X_test_data_accuracy)
```

```python
unseen_data = (7,107,74,0,0,29.6,254,31)
unseen_data_as_numpy_array = np.asarray(unseen_data)
unseen_data_reshape = unseen_data_as_numpy_array.reshape(1,-1)
std_unseen_data = scaler.transform(unseen_data_reshape)
predict = classifier .predict(std_unseen_data)
print(predict)
if predict[0] == 0:
  print("Person is non diabetic")
else:
  print("Person is diabetic")
```
